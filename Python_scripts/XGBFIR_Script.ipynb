```python
import pandas as pd
import numpy as np
import time
import xgboost as xgb
from sklearn.cross_validation import train_test_split
```

```python
start = time.time()
Data = pd.read_csv("/home/puneetj/PortoSegeo/train.csv")
print ('Time taken to read: ', time.time() - start)
del Data["id"]
```

```python
 Data2 = pd.get_dummies(Data, columns=[x for x in Data.columns if "_cat"  in x], drop_first = True)

```

```python
[x for x in Data.columns if "_cat"  in x]
```

```python

X_train, X_test, y_train, y_test = train_test_split( Data2[[x for x in Data2.columns if "target" not in x]], 
                                                    Data2["target"],
                                                    test_size=0.7, 
                                                    random_state=1000)    
```

```python
T_train_xgb = xgb.DMatrix(X_train,
                          y_train)
T_test_xgb = xgb.DMatrix(X_test, y_test)
watchlist  = [ (T_train_xgb,'train'),(T_test_xgb,'eval')]

params = {"objective": "reg:logistic",
          "booster":"gbtree",
    #      "eval_metric ":"mae",
          "max.depth" : "9",
         "min_child_weight" :"50",
         #"gamma":".005",
          "subsample":"0.3",
          "colsample_bytree": "0.6",
          "eta": ".03"
         }
gbm = xgb.train(dtrain =T_train_xgb,
                evals=watchlist,
                params=params,
                verbose_eval= True,
                #feval =rmsle_eval, 
                num_boost_round=1000,
          #      early_stopping_rounds=50
               )
```

```python
import xgbfir
xgbfir.saveXgbFI(gbm, 
                 feature_names=T_train_xgb.feature_names,
                 MaxTrees =1000,
                 MaxInteractionDepth =6,
                 OutputXlsxFile = 'Xgbfi_with_region.xlsx')
```
